```python
import numpy as np
import matplotlib.pyplot as plt
```

- 質点、バネ定数を $m, \ k$ とすると、固有各振動数は $ \omega_n = \frac{k}{m}$ となり、$x$ は以下の式で表すことができる
$$ x = A \sin \omega_n t + B \cos \omega_n t $$
- 初期条件を $ t=0 \rightarrow x = x_0, \dot{v_0} = v_0$
$$ x = \frac{v_0}{\omega_n} \sin \omega_n t + x_0 \cos \omega_n t $$

```python
m, k = 1, 1
x_0, v_0 = 0, 5
t = np.arange(0, 10, 0.1)


def single_vibration(m, k, x_0, v_0):
    omega_n = np.sqrt(k / m)
    A = v_0 / omega_n
    B = x_0
    x = A * np.sin(omega_n * t) + B * np.cos(omega_n * t)
    return omega_n, A, B, x


omega_n, A, B, x = single_vibration(m, k, x_0, v_0)

plt.figure(figsize=(6, 3))
plt.plot(t, x)
plt.title(
    f"$\omega_n={omega_n:.02f}, x={A:.02f} \sin \omega_n t + {B} \cos \omega_n t$"
)
plt.show()
```

```python
m, k = 1, 1
x_0, v_0 = 0, 5


def single_vibration(m, k, x_0, v_0):
    omega_n = np.sqrt(k / m)
    A = v_0 / omega_n
    B = x_0
    x = A * np.sin(omega_n * t) + B * np.cos(omega_n * t)
    return omega_n, A, B, x


t = np.arange(0, 10, 0.1)
initial_conditions = [
    [1, 1, 0, 5],
    [1, 1, 5, 0],
    [1, 1, 5, 5],
]

fig, axs = plt.subplots(
    1, len(initial_conditions), figsize=(len(initial_conditions) * 5, 3)
)
for ax, (m, k, x_0, v_0) in zip(axs, initial_conditions):
    omega_n, A, B, x = single_vibration(m, k, x_0, v_0)
    ax.plot(t, x)
    ax.set_title(f"$x={A:.02f} \sin {omega_n} t + {B} \cos {omega_n} t$")
plt.show()
```
